```python
import pandas as pd

# De praxe, carregamos o arquivo csv usando panda e definimos inicialmente o dataframe como df
df = pd.read_csv('SICOR_CONTRATOS_MUNICIPIO_2020_BRUTO.csv', header=0, sep=';', engine='python', on_bad_lines='skip', encoding='latin-1')
```

## ANALISE DOS DADOS FILTRADO POR ESTADO (CE)

```python
# Aqui nos apenas estamos filtrando o dataframe para aparecer somente os dados do estado do ceara, assim descartando os demais por não ser no nosso interesse
df_filtrado_CE_2020 = df[df['nomeUF'] == 'CE']
df_filtrado_CE_2020.to_csv('df_filtrado_CE_2020.csv', index=False)
display(df_filtrado_CE_2020.iloc[1:6])  # Exibe as linhas de 1 a 5 do dataframe filtrado para o estado do Ceará
```

## ANALISE COM FILTRO DO CEARA + PRONAF (COD = 1)

```python
df_filtrado_CE_2020_cod_pronaf = df_filtrado_CE_2020[df_filtrado_CE_2020['cdPrograma'] == 1]
display(df_filtrado_CE_2020_cod_pronaf.iloc[1:6])  # Exibe as linhas de 1 a 5 do dataframe filtrado para o estado do Ceará
df_filtrado_CE_2020_cod_pronaf.to_csv('df_filtrado_CE_2020_cod_pronaf.csv', index=False)
```
